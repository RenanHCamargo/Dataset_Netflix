As informações acima estão confusas, com ela não é possível definir por exemplo qual o genero que prevalece no Dataset, pois ele não está mostrando os generos separadamente.


```python
type(dados['genres'][1])
```

Com o retorno do comando, descobri que a tipagem está errada, o dado foi inserido como string, vou transformar o dado em lista porque no futuro será melhor de trabalhar.

```python
dados_v2 = dados_tratados.copy()
dados_v2['genres_typed'] = dados.apply(lambda x: eval(x['genres']), axis=1)
dados_v2['genres_typed']
```

```python
type(dados_v2['genres_typed'][1])
```

Com os dados tipados corretamente, é possível fazer a contagem correta de cada tipo de genero.

```python
from itertools import chain
lista = dados_v2['genres_typed'].to_list()
lista_com_todos_generos = list(chain.from_iterable(lista))
count_genres = pd.Series(lista_com_todos_generos).value_counts()
count_genres
```

```python
#Passando a coluna tratada para o Dataset tratado
dados_tratados['genres'] = dados_v2['genres_typed']
type(dados_tratados['genres'][1])
```
